```python
import os
from dotenv import load_dotenv
load_dotenv()

```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
## LANGHCAIN TRACKING 
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
# LLM MODEL 
from langchain_openai import ChatOpenAI
model =ChatOpenAI(model="gpt-4")

```

```python
# response from llm 
model.invoke("what is langtchain") 
```

```python
# chatprompttemplets
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system","give the answer correctly"),
        ("user","{input}")
    ]
)
```

```python
# prompt outputs
prompt.invoke({"inputs:what is genverative aii"})
```

```python
# chain 
chain =prompt|model
chain.invoke("inputs:what is gen ai")
```

```python
print("heloo")
```

```python
# stroutputs --
from langchain_core.output_parsers import StrOutputParser
outputs=StrOutputParser()
chain =prompt|model|outputs
res = chain.invoke("inputs:what is agentic ai ")
```
